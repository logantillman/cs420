```python
import numpy as np
import matplotlib.pyplot as plt

# Seed the random number generator
np.random.seed(0)
```

```python
def elementary_ca(steps, size, rho, rule_index):
    
    # Init x with size and time steps
    x = np.zeros((steps, size))
    
    # Set up x with initial cars
    x[0,:int(rho*size)] = 1
    np.random.shuffle(x[0,:])
    speeds = []
    
    lookup_table = [int(x) for x in reversed(bin(rule_index)[2:].zfill(8))]
    
    # Step through all time steps and update elements
    for t in range(steps - 1):
        for j in range(size):
            left = (j - 1) % size
            right = (j + 1) % size
            n_state = 4 * x[t, left] + 2 * x[t, right]
            x[t+1, j] = lookup_table[int(n_state)]
    
    return x
```

```python
steps = 100
size = 100
rho = 0.3

# Calculate x
x = elementary_ca(steps, size, rho, 107)

# Plot it
plt.imshow(x, cmap='binary')
```
