```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
```

```python
def elementary_ca(steps, size, rho, rule_index):
    # Setup x with all zeroes
    x = np.zeros((steps, size))
    
    # Initialize the first state of x with 1's as appropriate 
    x[0,:int(rho*size)] = 1
    np.random.shuffle(x[0,:])
    
    # EXTRACT RULE TABLE!
    lookup_table = [int(x) for x in reversed(bin(rule_index)[2:].zfill(8))]
    
    # Loop through and update 
    for t in range(steps-1):
        for i in range(size):
            left = (i-1)%size
            right = (i+1)%size
            n_state = 4*x[t,left]+2*x[t,i]+x[t,right]
            x[t+1,i] = lookup_table[int(n_state)]
            
    return x
```

```python
steps = 100
size = 100
rho = 0.3

x = elementary_ca(steps, size, rho, 30)
plt.imshow(x,cmap='binary')
```
